```python
import requests
import pandas as pd
# URL API
url = 'https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi'

# Mengirim GET request
response = requests.get(url)

# Mengecek apakah request berhasil (kode status 200)
if response.status_code == 200:
    # Mengonversi respons ke format JSON
    data = response.json()

    # Mengakses data sesuai kebutuhan
    # Contoh: mencetak judul dari artikel pertama
    # print(data[0]['title'])
    # Mengambil nilai dari kunci 'data'
    data_value = data.get('data', [])

    # Membuat DataFrame dari data_value
    df = pd.DataFrame(data_value)
    print(df.head())

else:
    print('failed get data:', response.status_code)
```

```python
df.dtypes
```

```python
df['isoDate'] = pd.to_datetime(df['isoDate'])
```

```python
df['creator'] = df['link'].str.replace(r'https:\/\/www\..*?\/', '')
```

```python
df.head()
```

```python
aggregated = df.groupby('creator').sum().reset_index()
```

```python
sorted = aggregated.sort_values(by='creator', ascending=False)

# Print DataFrame hasil aggregasi dan pengurutan
print(sorted)
```
